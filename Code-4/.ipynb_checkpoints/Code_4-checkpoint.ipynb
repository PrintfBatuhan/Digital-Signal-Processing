Basit kesirlere ayırma yöntemiyle ters z dönüşümü

```python
# gerekli kütüphanelerin eklenmesi
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sgnl
```

```python
z=np.array([0])             #verilen sinyalin sıfırlarının yazılması
p=np.array([1/4,1/2])       #verilen sinyalin kutuplarının yazılması
b, a = sgnl.zpk2tf(z, p, 1) # kutup ve sıfırların transfer fonksiyonuyla birleştirilmesi pay ve payda şeklinde çıktı almış olacağız.
print(b,a)
```

```python
#payda kısmının oluşturulmasını bu şekilde de bulabiliriz
a=sgnl.convolve(np.array([1,-1/4]),np.array([1, -1/2])) 
print(a)
```

```python
r, p, k = sgnl.residuez(b,a)
print(r)   #terimlerin pay kısmı
print(p)   # sistemin kutupları
print(k)
```

```python
poless = np.roots(a) #paydanın polinom katsayıları
print(poless)
```

## Ön çalışma 10) 

Z dönüşümü verilen sistemin frekans cevabının çizilmesi ve ardından giriş işaretinin bu sisteme uygulanması

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sgnl
%matplotlib notebook
```

```python
# sistemin sıfır ve kutuplarla oluşturulması
zeros=np.array([-1])
poles=np.array([1j/2,-1j/2,1.0/4])

w,H=sgnl.freqz_zpk(zeros,poles,1)

plt.plot(w/np.pi, abs(H))
plt.title('$H(Z)$ nin Frekans cevabı')
plt.ylabel('Magnitude')
plt.xlabel('$\omega$ x $\pi$ (rad/sample)')
plt.grid()
plt.show()
```

```python
#sistemi transfer fonksiyonu yağısında tekrar elde edelim
num, denum = sgnl.zpk2tf(zeros, poles, 1)

w1, H_tf = sgnl.freqz(num,denum)

plt.figure()
plt.plot(w1/np.pi, abs(H_tf))

plt.title('Frequency Response of $H(Z)$')
plt.ylabel('magnitude')
plt.xlabel('$\omega$ x $\pi$ (rad/sample)')
plt.grid()
plt.show()
```

```python
#Şimdi, giriş sinyalini tanımlayacağız ve verilen sisteme uygulayacağız.
n= np.arange (0,40,1)
xn=2*np.cos(0.2*np.pi*n) + np.sin(0.9*np.pi*n)

yn=sgnl.lfilter(num, denum, xn)
plt.figure()
plt.stem(n,yn,use_line_collection=True)
plt.grid()
plt.show()
```

## ÖN HAZIRLIK SORULARI (Ödev)

## 1) 
### a) Sistem fonksiyonu H(z)'nin bulması

 𝑦[𝑛]=1/2𝑦[𝑛−1]−4𝑦[𝑛−2]+2𝑦[𝑛−3]+𝑥[𝑛]+2/3𝑥[𝑛−1]+1/9𝑥[𝑛−2]
şeklinde verilen DZD fark denkleminde

a) Sistem fonksiyonu H(z)'yi bulmak için verilen fark denkleminin z-dönüşümü'nü alırsak :

    Y(Z)=(1/2*Z^-1)Y(Z)-(4*Z^-2)Y(Z)+(2*Z^-3)Y(Z)+X(Z)+(2/3*Z^-1)X(Z)+(1/9*Z^-2)X(Z)
    Y(z) ve X(Z) parantezlerine alırsak:
        Y(Z)(1-1/2Z^-1 + 4Z^-2 - 2Z^-3)=X(Z)(1+2/3Z^-1 + 1/9Z^-2) 
   Biz Y(Z)=X(Z)H(Z) eşitliğinin olduğunu biliyoruz. Bu eşitliği düzenlersek H(Z)=Y(Z)/X(Z) dir. Elde ettiğimiz denklemi   düzenlersek:
    
     H(Z)=Y(Z)/X(Z) = (1+2/3Z^-1 + 1/9Z^-2)/(1-1/2Z^-1 + 4Z^-2 - 2Z^-3) bulunur. bu denklemi çarpanlarına ayırırsak;
     H(Z)= (1+ 1/3Z^-1)(1+ 1/3Z^-1)/(1+4Z^-2)(1-1/2Z^-1)
     H(Z)= (1+ 1/3Z^-1)(1+ 1/3Z^-1)/(1+j2Z^-1)(1-j2Z^-1)(1-1/2Z^-1) elde etmiş oluruz. 
 Bu şekilde işlemlerimizi daha rahat bir şekilde gerçekleştireceğiz.



### b) Kutup-sıfır diyagramının çizdirilmesi

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sgnl
import zplane
```

```python
#verilen sinyalin sıfırlarının yazılması
zeross=np.array([-1/3,-1/3])
#verilen sinyalin kutuplarının yazılması
poless = np.array([-2j,2j,1/2])
k=1
# kutup ve sıfırların transfer fonksiyonuyla sistemi oluşturması
b, a=sgnl.zpk2tf(zeross,poless,k) 
#sistemin kutup sıfır diyagramı
zplane.zplane(b,a)
```

### c) ROC bölgesini şu iki durum için inceleyelim


    c.1.durum: h[n]=0 ve n<0 için;
      Bu durum sistemin nedensel olma koşuludur. Bu durum sistemimizin sağ yanlı olduğunu söyler.
      ROC'un özelliklerinden ROC bölgesinin içinde kutup bulunduramayacağını biliyoruz. Dolayısıyla sistemin nedensel/sağ yanlı olması için en dış kutbu ROC bölgesi seçmemiz gerekir böylece içinde kutup bulundurmaz. 
Bulduğumuz kutup-sıfır diyagramı üzerinde bu koşulun salanması için;
      ROC: |z|>2j olması gerekir.
  
    c.2.durum: sistemin Fourier dönüşümü varsa;
      Bu durum sistemin kararlı olma koşuludur. Sistemin Fourier dönüşümü varsa ROC birim çemberi içermelidir. Bu koşul sağlanırsa sistemimiz kararlıdır diyebiliriz. Bulduğumuz kutup-sıfır diyagramı üzerinde bu koşulun sağlanması için ROC: |z|>0.5 olması gerekir. Fakat ROC bölgesinin içinde kutup bulunduramayacağını biliyoruz dolayısıyla |z|<2j den küçük olmak zorunda kalır. 
Bu koşulun sağlanması için;
  ROC: 0.5<|z|<2j  olması gerekir.


### d) c.1 durumunun sağlanması

 c.1 durumunun sağlanması için gereken koşulları yukarıda yazdık yani sistemimizin sağ yanlı olduğunu ve  ROC: |z|>2j olması gerektiğini söyledik.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sgnl
%matplotlib notebook
```

```python
#kuvvet serisi yöntemiyle
num=np.array([1, 2.0/3, 1.0/9, 0])
denum=np.array([1, -1.0/2, 4.0, -2.0])

n, x=sgnl.dimpulse((num, denum,1),x0=0,n=10)


print(np.squeeze(x))
```

```python
plt.stem(n,np.squeeze(x),use_line_collection=True)
plt.grid()
```

Çizdirdiğimiz dürtü cevabından yola çıkarak görüyoruz ki sistem nedensel fakat kararlı değil çünkü sonsuza doğru gidiyor sınırlandığı bir değer yok.

### e) Sistemin aynı anda hem kararlı hem nedensel olması:

 Bu sistem aynı anda hem kararlı hem de nedensel OLAMAZ. Bunun nedeni aslında c) kısmında c.1 ve c.2 durumlarını incelerken belirttik. 
     
     Bu sistemin nedensel olması için ROC: |z|>2j , kararlı olabilmesi için  0.5<|z|<2j olması gerekiyor. Bu yüzden bu aralıklarda koşulu sağlayan ortak bir ROC bulamıyoruz. 
     
Sistemimizin en kararlı hem nedensel olması demek birim çemberi içermeli ve sağ yanlı olmasını gerektirir. 
    
    Bu koşuşu sağlayabilecek sınır ise ROC: |z|>0.5 tir.

ROC içinde kutup içeremeyeceğinden dolayı ROC:|z|>0.5 koşulunun sağlanması yani hem nedensel hem kararlı olabilmesi için 2j ve -2j kutuplarının olmaması gerekir. Bu sistemi bu şekilde düzenlemek istersek. 

     H(Z)= (1+ 1/3Z^-1)(1+ 1/3Z^-1)/(1+4Z^-2)(1-1/2Z^-1) gördüğümüz gibi komplex kutupları veren (1+4Z^-2) değerini buradan yok etmemiz lazım
     
    H(Z) değerimizi (1+4z^-2) ile çarparsak komplex kutuplardan kurtulmuş oluruz ve yeni bulduğumuz sistem fonksiyonumuz 
    H(Z)'=(1+1/3z^-1)(1+1/3z^-1)/(1-1/2z^-1) şeklinde olur.
    Gördüğümüz gibi bu sistem ROC:|z|>0.5 olduğunda istenilen hem kararlı hem de nedensel durumunu sağlar.

### f) (e) şıkkında bulduğunuz sistemin kutup-sıfır diyagramı ve dürtü cevabını bulunuz ve çizdiriniz 

Kutup-sıfır diyagramının çizdirilmesi

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sgnl
import zplane
```

```python
zeross=np.array([-1/3,-1/3])
poless = np.array([1/2])
k=1

b, a=sgnl.zpk2tf(zeross,poless,k)
zplane.zplane(b,a)
```

Dürtü cevabının çizdirilmesi

```python
num=np.array([1, 2.0/3, 1.0/9, 0])
denum=np.array([1, -1.0/2, 0, 0])

n, x=sgnl.dimpulse((num, denum,1),x0=0,n=10)

print(np.squeeze(x))
```

```python
plt.figure()
plt.stem(n,np.squeeze(x),use_line_collection=True)
plt.grid()
```

Kutup-sıfır diyagramında görüldüğü gibi imajiner eksendeki kutupları yok ettik. Sistemimizin sağ yanlı olursa nedensellik koşulu sağlanmış olur. Bu durum reel kısımdaki 1/2 kutbumuzun dışındaki alanların taranması gerektiğini söyler. Kutbumuz 1 den küçük olduğu içinde birim çemberi içermiş olur. Birim çemberi içermiş olması fourier dönüşümünün olduğu ve kararlı olduğu göstergesidir. Bu durum dürtü cevabı üzerindeki etkisi 1 noktasında dürtü cevabımız en yüksek değerini alıyor yani birim çember noktasında. Sonrada azalarak sistemin sıfır değerine yakınsamasına neden oluyor yani kararlılık sağlanıyor.

### g) (e) şıkkında bulunan sistemin kutup-sıfır diyagramına bakarak, kesim frekansını (cut-off) belirlenmesi

```python

w = np.linspace(0, 2*np.pi, 100)
Hw = (1+1/3*np.exp(-1j*w))/(1-(1/2)*np.exp(-1j*w))

plt.figure()
plt.plot(w/np.pi, abs(Hw))
plt.xlim(0,2)
plt.xlabel('$\omega$ x$\pi$ rad/sample')
plt.grid()
```

Ön çalışma örnek 11 de verilen örnekten farklı olarak bizim elde ettiğimz sistemin kutup sıfır diyagramında görüyoruz ki sıfırlar çift kutupludur. Tanım gereği, sistemin “kutbunun” payda teriminin bir köküdür. Bu nedenle, bir kutup sistemin büyüklüğünü artırmaya çalışırken , 'sıfır'  büyüklüğü azaltmak ister.
Birim çemberinde z = j'de sıfıra doğru ilerledikçe, büyüklük azalır. Sistem sıfırında, z = j, frekans yanıtı beklendiği gibi sıfırdır. Kutup gösterimlerinde, z = j = ejπ / 2, bu nedenle, pozitif gerçek eksenden sıfırın açısı, sistemin kesme frekansıdır.

z = j = ejπ / 2 bu kısım hatalı, sistemin sıfırı z=-1'de olması gerekiyor. Geri kalan kısımlarda sorun yok, sadece ikisi arasındaki tutarsızlığın farkında olmanızı istedim.
