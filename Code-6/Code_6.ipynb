Down sampling (Seyrek Örnekleme)

```python
#gerekli kütüphanelerin çağırılması
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#x(t) işaretimizin tanımlanması

F=100                      #x(t)nin temel frekansı
T=1/F                      # temel periyodu
t=np.arange(0,2*T,0.0001)  #iki periyotluk t zaman indisi tanımlanması

x=5*np.cos(200*np.pi*t)  #x(t) işaretimiz
```

```python
# x(t) işaretimizin Fs=1200Hz ile örneklenmesi ile x[n] işaretimizin elde edilmesi

Fs=1200                    #örnekleme frekansımız
Ts=1/Fs                    #örnekleme periyodumuz
N=12                       # bir periyotluk örnek sayısı
n=np.arange(0,2*N)         #iki periyotluk aralık belirlenmesi

xn=(5*np.cos(200*np.pi*n*Ts)) #örneklenmiş işaretimiz=x[n]
```

```python
M=2     #Downsampling oranı

xn_d=xn[np.arange(0, np.size(xn,0),M)]  #x[n]in M katlarındaki indislerindeki örneklerin bulunması
Nn_d=len(xn_d)         #uzunluğun bulunması
n_d=np.arange(0,Nn_d)  #indis aralıklarının belirlenmesi
```

```python
plt.figure(figsize=(8,10))

plt.subplot(311)        #x(t)nin çizdirilmesi
plt.plot(t,x)
plt.xlabel("$t(sn)$")
plt.ylabel("$x(t)$")
plt.grid()

plt.subplot(312)        #x(t)nin Fs=1200Hz ile örneklenmiş hali X[n] çizdilmesi
plt.stem(n,xn,use_line_collection=True)
plt.xlabel("$n (örnek)$")
plt.ylabel("$x[n]$")
plt.grid()

plt.subplot(313)        #seyrek örneklenmiş x[n]in (xn_d[n]) çizdirilmesi
plt.stem(n_d, xn_d,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x_d[n]$")
plt.grid()
plt.show()
```

```python
#x(t) işaretimizin Fs=600Hz ile örneklenmesi sonucu elde edilen x2[n] işaretimiz
Fs=600       #Örnekleme frekansı
Ts=1/Fs      #örnekleme periyodu
N=6          #bir periyottaki örnek sayımız
n=np.arange(0,2*N)   #iki periyotluk çizdirmek için aralığın tanımlanması

xn2=5*np.cos(200*np.pi*n*Ts)  #örneklenmiş x2[n]işaretimizin tanımlanması

plt.figure(figsize=(8,4))     #işaretimizin çizdirilmesi
plt.stem(n,xn2,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x_2[n]$")
plt.grid()
plt.show()
```

```python

M=4    #Downsampling oranı

xn_d=xn[np.arange(0, np.size(xn,0),M)]  #x[n]in M katlarındaki indislerindeki örneklerin bulunması
Nn_d=len(xn_d)         #uzunluğun bulunması
n_d=np.arange(0,Nn_d)  #indis aralıklarının belirlenmesi

plt.figure()       #seyrek örneklenmiş x[n]in (xn_d[n]) çizdirilmesi
plt.stem(n_d, xn_d,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x_d[n]$")
plt.grid()
plt.show()
```

x[n] işaretinin M=2 yerine M=4 ile seyrek örnekleme yapılması sonucunda yukarıdaki grafikten görüldüğü üzere x(t) işaretini geri elde edemeyeceğimizi anlıyoruz. x(t) işaretinin temel frekansı Fn=200 Hz ve M=4 ile seyrek örneklenen işaretin örnekleme frekansı ise Fs=300 Hz’dir. Örnekleme frekansı, işaretin temel frekansının iki katından küçüktür, dolayısıyla Nyquist şartını sağlamamaktadır.

### Örnek 2)

Seyrek örneklemenin frekans domaininde incelenmesi

```python
#gerekli kütüphanelerin çağrılması
from matplotlib import pyplot as plt
import numpy as np
```

```python
N=80   #örnek sayımız
nTs=np.arange(-10,10,20/N)  #nTs indis tanımız
xn=np.sinc(nTs)**2    #x(nTs) işaretinin tanımlanması
n=np.arange(-40,40)   #x[n]in indislerinin tanımlanması

# x[n]in Fourier Transformunun alınması
w=np.arange(-np.pi,np.pi, 2*np.pi/N)
xw=np.fft.fftshift(np.fft.fft(xn,N)/N)
```

```python
# M=2 ile seyrek örneklenmenin yapılması

M_1=2 #Seyrek örnekleme oranı
xn_d1=xn[np.arange(0,np.size(xn,0),M_1)] # x[n]in M=2 katlarındaki örneklerin alınması 
N_d1=(round)(N/M_1)
n_d1=np.arange(-N_d1/2,N_d1/2) # xn_d1[n] işaretimizin indis aralığı

# M=2 ile örneklenmiş işaretimizin (xn_d1[n]) Fourier Transformunun alınması ile oluşan işaretimiz= Xd_1(ω)
w_d1=np.arange(-np.pi,np.pi,2*np.pi/N_d1)
xw_d1=np.fft.fftshift(np.fft.fft(xn_d1,N_d1)/N_d1)
```

```python
#M=3 ile seyrek örneklenmenin yapılması
M_2=3
xn_d2=xn[np.arange(0,np.size(xn,0),M_2)] # x[n]in M=3 katlarındaki örneklerin alınması 
N_d2=(round)(N/M_2)
n_d2=np.arange(-N_d2/2,N_d2/2) # xn_d2[n] işaretimizin indis aralığı

# M=3 ile örneklenmiş işaretimizin (xn_d2[n]) Fourier Transformunun alınması ile oluşan işaretimiz= Xd_2(ω)
w_d2=np.arange(-np.pi,np.pi,2*np.pi/N_d2)
xw_d2=np.fft.fftshift(np.fft.fft(xn_d2,N_d2)/N_d2)
```

```python
plt.figure(figsize=(9,15))

# x[n] işaretimizin çizdirilmesi
plt.subplot(321)
plt.stem(n,xn,use_line_collection=True) 
plt.xlabel("$n (örnek)$") 
plt.title("$x[n]$") 
plt.grid()

# x[n]in Fourier Transformu olan X(ω) işaretinin çizdirilmesi
plt.subplot(322)
plt.stem(w/np.pi,abs(xw),use_line_collection=True) 
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X(\omega)$")
plt.grid()

# x[n] işaretimizin M=2 ile seyrek örneklenmiş hali xd1[n] in çizdirilmesi
plt.subplot(323)
plt.stem(n_d1,xn_d1,use_line_collection=True) 
plt.ylabel("M=2 için")
plt.xlabel("$n (örnek)$") 
plt.title("$x_d1[n]$") 
plt.grid()

# xd1[n] in Fourier Dönüşümü olan Xd1(ω) in çizdirilmesi
plt.subplot(324)
plt.stem(w_d1/np.pi,abs(xw_d1),use_line_collection=True) 
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X_d1(\omega)$")
plt.grid()

# x[n] işaretimizin M=3 ile seyrek örneklenmiş hali xd2[n] in çizdirilmesi
plt.subplot(325)
plt.stem(n_d2,xn_d2,use_line_collection=True) 
plt.ylabel("M=3 için")
plt.xlabel("$n (örnek)$") 
plt.title("$x_d2[n]$") 
plt.grid()

# xd2[n] in Fourier Dönüşümü olan Xd2(ω) in çizdirilmesi
plt.subplot(326)
plt.stem(w_d2/np.pi,abs(xw_d2),use_line_collection=True) 
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X_d2(\omega)$")
plt.grid()
plt.show()
```

### Örnek 3)

```python
#gerekli kütüphanelerin çağırılması
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
Fs=1200           #örnekleme frekansımız
Ts=1/Fs           #örnekleme periyodumuz
N=12              #örnek sayımız
n=np.arange(0,N)  #bir periyot için örnekleme indisimiz
xn=5*np.cos(200*np.pi*n*Ts)  #x[n] işaretimizin tanımlanması
```

```python
# sık örneklemenin yapılması
L=2     #sık örnekleme katsayısı
Nn_u=N*L   #sık örneklenmiş işaretin uzunluğu
xn_u=np.zeros(Nn_u)  #işaretin uzunluğu kadar sıfır dizisi oluşturulması
xn_u[np.arange(0,len(xn_u),L)]=xn  #oluşturulan sıfır dizisinin üzerine x[n] değerlerinin atanması

n_u=np.arange(0,Nn_u)
```

```python
#lineer interpolasyon işlemi

hn=np.array([0,1/2,1,1/2,0])  #L=2 için h[n] işaretinin tanımlanması

xn_i=np.convolve(xn_u,hn,'full') #konvolüsyon işleminin yapılması
n_i=np.arange(0,len(xn_i))       # indis değerimiz
```

```python
#grafiklerin çizdirilmesi
plt.figure(figsize=(8,12))

#x[n] değerinin çizdirilmesi
plt.subplot(311)   
plt.stem(n,xn,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x[n]$")
plt.grid()

#x[n] değerinin aralarına L=2 aralıklarla 0 eklenmesiyle oluşturulan x_u[n] işaretinin çizdirilmesi
plt.subplot(312)
plt.stem(n_u,xn_u,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x_u[n]$")
plt.grid()

#x_u[n]işaretinin interpolasyon sonucu sık örneklenmiş işaretimiz x_i[n] çizdirilmesi
plt.subplot(313)
plt.stem(n_i,xn_i,use_line_collection=True)
plt.xlabel("$n(örnek)$")
plt.ylabel("$x_i[n]$")
plt.grid()
plt.show()
```

YORUM:

           x[n] işareti L=2 ile sık örnekleme yaptığımızda görüyoruzki x[n] işaretimizin aralarına 2-1=1 tane sıfır konarak x_u[n] elde ettik. 
    
           h[n] işaretimizi L=2 için array şeklinde bir üçgen dürtü olarak tanımladık. 
    
           x_u[n], h[n] ile lineer interpolasyon işlemine soktuk bu işlem x_u[n] ve h[n]in konvolüsyon işlemidir. Lineer interpolasyon işleminin, bulduğumuz x_u[n] işaretindeki örnekleri doğrusal bir şekilde birleştiriyor gibi gösterdiğini söyleyebiliriz.
    
    Bu işlem sonucunda x[n]in L=2 ile sık örneklenmiş hali x_i[n] işaretimizi elde ettik.
   


### Örnek 4)

Sık örneklemenin frekans domaininde incelenmesi

```python
#gerekli kütüphanelerin çağrılması
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
N=40   #örnek sayımız
nTs=np.arange(-10,10,20/40)  #nTs aralığının belirlenmesi
n=np.arange(-20,20)          #x[n] işaretimizin indis değerleri
xn=np.sinc(nTs)**2           #x(nTs) işaretinin tanımlanması

#x[n] işaretinin fourier transformunun alınması
w=np.arange(-np.pi,np.pi,2*np.pi/N)
xw=np.fft.fftshift(np.fft.fft(xn,N)/N)
```

```python
#sık örnekleme işleminin yapılması
L=2               #örnekleme katsayımızın belirlenmesi
N_u=N*L           #Sık örnekleme sonucu oluşacak işaretimizin uzunluğu
xn_u=np.zeros(N_u)  # oluşacak işaretimizin boyu kadar 0 dizisi oluşturulması
xn_u[np.arange(0,len(xn_u),L)]=xn #oluşturulan 0 dizisinin her L eleman katına x[n] değerinin atanması
n_u=np.arange(-N_u/2,N_u/2)

#x_u[n] işaretinin fourier transformunun alınması
w_u=np.arange(-(np.pi),(np.pi),2*(np.pi)/N_u)
xw_u=np.fft.fftshift(np.fft.fft(xn_u,N_u)/N_u)
```

```python
# AGF (alçak geçiren filtre) ile frekans cevabımızın oluşturulması

hw=np.zeros(N_u) #sıfır vektörünün oluşturulması

w_i=np.arange(-(np.pi),(np.pi),2*(np.pi)/N_u)
indis=np.where(abs(w_i)<=np.pi/2)  #kesim frekansı (W_c) pi/2 olan filtrenin pi/2 den küçük değerlerinin bulunması
hw[indis]=L   #filtre kazancı =L kadar değer atanması


#X_i(w) işaretinin tanımlanması
xw_i=xw_u*hw  #frekans uzayınca çarpım konvolüsyona karşılık geleceği için çarpıyoruz
N_i=len(xw_i)
```

```python
# grafiklerin çizdirilmesi
plt.figure(figsize=(9,15))  #grafik boyutunun ayarlanması

#x[n] işaretinin çizdirilmesi
plt.subplot(321)
plt.stem(n,xn,use_line_collection=True)
plt.xlabel("n (örnek)")
plt.title("$x[n]$") 
plt.grid()

#X(W) işaretinin çizdirilmesi
plt.subplot(322)
plt.stem(w/np.pi,abs(xw),use_line_collection=True)
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X(\omega)$ (Genlik)")
plt.grid()

#x_u[n] işaretinin çizdirilmesi
plt.subplot(323)
plt.stem(n_u,xn_u,use_line_collection=True) 
plt.xlabel("n (örnek)") 
plt.title("$x_u[n]$") 
plt.grid()

#X(W) işaretinin çizdirilmesi
plt.subplot(324)
plt.stem(w_u/np.pi,abs(xw_u),use_line_collection=True) 
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X_u(\omega)$ (Genlik)") 
plt.grid()

#x_i[n] işaretinin çizdirilmesi
plt.subplot(325)
plt.stem(w_i/np.pi, abs(hw),use_line_collection=True)
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$H(\omega)$") 
plt.grid()

#X(W) işaretinin çizdirilmesi
plt.subplot(326)
plt.stem(w_i/np.pi,abs(xw_i),use_line_collection=True) 
plt.xlabel("$\omega$ / $\pi$") 
plt.title("$X_i(\omega) (Genlik)$")
plt.grid()
plt.show()
```

İşaretimizi zaman domaininde inceleyerek örneklenip örneklenmediğini kontrol edelim

```python
#X_İ(W) işaretinin ters fourier dönüşümü ile x_i[n] işaretinin elde edilmesi

xn_i=np.fft.ifft(np.fft.ifftshift(xw_i),N_i)*N_i
ni=np.arange(-N_i/2,N_i/2)

plt.figure(figsize=(9,4))
plt.stem(ni,abs(xn_i),use_line_collection=True)
plt.xlabel("n(örnek)")
plt.title("$x_i[n]$")
plt.grid()
plt.show()
```

YORUM:
        
      İlk önce x[n] işaretini ve X(w) işaretlerimizi oluşturduk ardından L=2 ile seyrek örnekleme işlemine başladık. İlk olarak örneklerimizin arasına L-1=2-1=1 tane 0 koyarak x_u[n] işaretimizi elde ettik. X_u(W) işaretimizi incelediğimizde ise X(W) işaretinin temel periyodunda bir tane üçgen darbe formu varken, X_u(w) işaretimizde iki adet üçgen darbe işareti bulunuyor. 

      Dolayısıyla H(W)= AGF (alçak geçiren filtre) ile interpolasyon işlemine soktuk. İşaretimizin korunması için filtre kazancımızı L kesim frekansımızı ise pi/L olarak seçtik. İnterpolasyon işlemimiz (koncolüsyon) frekans domaininde X_u(W)*H(W) şeklinde olacağı için bu oluşturduğumuz iki grafiği nokta çarpımı ile çarptığımızda beklenen X_i(W) işaretimizi elde ettik.
    
      Bu işaretimizin ters fourier dönüşümü aldığımızda x_i[n] işaretimizi elde ettik ve x[n] işaretimizle karşılaştırdığımızda görüyoruz ki x_i[n] işareti x[n]in örneklerinin arasında yeni indislerin bulunmuş halidir yani sık örneklenmiş halidir.

### a-Şıkkı)

Örnekleme frekansı 40kHz olan x[n] işaretinin örnekleme frekansının 50kHz olması isteniyor. Biz biliyoruz ki 
       
       SIK ÖRNEKLEME için x[n] giriş işareti ile xi[n] çıkış işaretimizin  örnekleme frekansları arasında Fs=FsL şeklinde   bir bağıntı var.

Eğer bize verilen blok diyagramındaki gibi ilk önce SIK ÖRNEKLEME sonrasında ise oluşan işaretimize SEYREK ÖRNEKLEME yapılması isteniyorsa; 

    Giriş örnekleme frekansımız ile çıkış örnekleme frekansımız arasında Fs'=FsL/M şeklinde bir bağıntı olacağınız söyleyebiliriz.
 Bu bilgiler ışığıda sorumuza dönersek. Örnekleme frekansı Fs=40kHz olan x[n] işaretinin örnekleme frekansı Fs'=50kHz olması istenirse yukarıda elde ettiğimiz bağıntıyı kullanarak hesaplayalım;

    50kHz=40kHzL/M  ise M50kHz=L40kHz ve 5M=4L eşitliği bulunur. Buradan da M=4, L=5 seçilmesi gerektiğini söyleybiliriz.

### b-Şıkkı)

```python
#Gerekli kütüphanelerin çağrılması
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#x[n] fonksiyonumuzun tanımlanması

N=20                 #istenilen örnek sayımız            
n=np.arange(0,N)     #indis değerimizin belirlenmesi
xn=np.cos(np.pi*n/5) #x[n] işaretimizin tanımlanması

#x[n] işaretimizin fourier transformunun alınması
w=np.arange(-np.pi,np.pi,2*np.pi/N)
xw=np.fft.fftshift(np.fft.fft(xn,N)/N)
```

Sık örnekleme işlemimiz:

```python
L=5          #örnekleme katsayımız
Nn_u=N*L     #yeni oluşacak işaretimizin boyutunun bulunması

#x_u[n]işaretimizin tanımlanması
xn_u=np.zeros(Nn_u)                #oluşan işaretimiz boyunca sıfır indisi tanımlanması
xn_u[np.arange(0,len(xn_u),L)]=xn  #tanımlanan sıfır indislerine L katlar boyunca x[n]işaretinin koyulması

n_u=np.arange(0,Nn_u)  #çizdireceğimiz işaretimizin aralığının oluşturulması

#x_u[n]işaretimizin fourier transformunun alınması
w_u=np.arange(-(np.pi),(np.pi),2*(np.pi)/Nn_u)
xw_u=np.fft.fftshift(np.fft.fft(xn_u,Nn_u)/Nn_u)
```

 Lineer interpolasyon işleminin yapılması

```python
#h[n] dürtü cevabımızın L=5 için belirlenmesi
hn=np.array([0,1/5,2/5,3/5,4/5,1,4/5,3/5,2/5,1/5,0])

#x_i[n] işaretimizin oluşturulması
xn_i=np.convolve(xn_u,hn,'same') #h[n] ile x_u[n] işaretimizin konvolüsyon işlemi ile x_i[n] işaretimizin elde edilmesi
n_i=np.arange(0,len(xn_i))

#x_i[n] işaretimizin fourier transformunun alınması
w_i=np.arange(-(np.pi),(np.pi),2*(np.pi)/Nn_u)
xw_i=np.fft.fftshift(np.fft.fft(xn_i,Nn_u)/Nn_u)
```

Seyrek örnekleme işlemimiz:

```python
M=4     #seyrek örnekleme oranımız

xn_d=xn_i[np.arange(0, np.size(xn_i,0),M)]  #x_i[n]in M katlarındaki indislerindeki örneklerin bulunması
Nn_d=len(xn_d)         #uzunluğun bulunması
n_d=np.arange(0,Nn_d)  #indis aralıklarının belirlenmesi

#x_d[n] işaretimizin fourier transormunun alınması
w_d=np.arange(-np.pi,np.pi,2*np.pi/Nn_d)
xw_d=np.fft.fftshift(np.fft.fft(xn_d,Nn_d)/Nn_d)
```

```python
#gerekli kütüphanelerin çağrılması
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#x[n] işaretimizin tanımlanması
N=20
n=np.arange(0,N)
xn=np.cos(np.pi*n/5)

#x[n] işaretimizin Fourier Transformu
w=np.arange(-np.pi,np.pi,2*np.pi/N)
xw=np.fft.fftshift(np.fft.fft(xn,N)/N)
```

Sık örnekleme işlemimiz

```python
L=5
Nn_u=N*L
xn_u=np.zeros(Nn_u)
xn_u[np.arange(0,len(xn_u),L)]=xn

n_u=np.arange(0,Nn_u)

#x_u[n] işaretimizin fourier transformu
w_u=np.arange(-(np.pi),(np.pi),2*(np.pi)/Nn_u)
xw_u=np.fft.fftshift(np.fft.fft(xn_u,Nn_u)/Nn_u)
```

AGF filtre tasarımımız:

AGFnin ideal olması zaman domainindeki filtre dürtü cevabı olan h[n], sinc() işareti formunda olmaktadır. Dolayısıyla frekans
domaininde ideal AGF uygulanırken, zaman domaininde x[n] işaretimizi sinc() işareti ile konvolüsyon yapıyoruz.

```python
#h[n] işaretimizin sinc fonksiyonu olarak tanımlanması
n_uTs=np.arange(-10,10,20/100)
hn=np.sinc(n_uTs)

#x_i[n] işaretimizin çizdirilmesi
xn_i=np.convolve(xn_u,hn,'same')
n_i=np.arange(0,len(xn_i))

#x_i[n] işaretimizin fourier transformunun alınması ve ardından AGF ile frekans domaininde çarpılması
xw_i=np.fft.fftshift(np.fft.fft(xn_i,Nn_u)/Nn_u)

#AGF filtremizin oluşturulması
hw = np.zeros(Nn_u) 
w_i = np.arange(-(np.pi), (np.pi), 2*(np.pi)/Nn_u) 
indis = np.where(abs(w_i)<=np.pi/L)                #kesim frekansı olarak pi/L alınmıştır
hw[indis] = L                                      #kazancımız L olarak tanımlanmıştır
xw_i = xw_u*hw     #frekans uzayınca konvolüsyon nokta çarpımına eşittir.
N_i=len(xw_i)
```

Seyrek örnekleme işlemimiz:

```python
M=4     

#xn_d[n] işaretimizin tanımlanması
xn_d=xn_i[np.arange(0, np.size(xn_i,0),M)]  
Nn_d=len(xn_d)        
n_d=np.arange(0,Nn_d)  

#xn_d[n] işaretimizin Fourier dönüşümü
w_d=np.arange(-np.pi,np.pi,2*np.pi/Nn_d)
xw_d=np.fft.fftshift(np.fft.fft(xn_d,Nn_d)/Nn_d)
```

Dolayısıyla x_i[n] işaretimiz b-şıkkında yapılan lineer interpolasyon ile bulunan x_i[n]den farklıdır. Burada bulduğumuz x_i[n] işaretimiz doğrusal değil daha çok oval/parabolik biçimde orijinaline yakın bulunmuştur. Bunun sebebi AGFnin ideal olması zaman domainindeki filtre dürtü cevabı olan h[n], sinc() işareti formunda olmaktadır. Dolayısıyla frekans domaininde ideal AGF uygulanırken, zaman domaininde x[n] işaretimizi sinc() işareti ile konvolüsyon yapmamızdır. Tamamen orijinal işaretimizin çıkmamasının sebebi seyrek örneklenmiş x_u[n] işaretimizin kenarlarında bulunan örnek sayılarımızın yetersiz olmasından kaynaklanmaktadır. Ayrıca sinc() fonksiyonumuz sonsuz uzunlukta sürekli zamanlı bir fonksiyondur biz bunu örneklerle, sınırlı bir şekilde ifade ettiğimizden dolayı çıktı işaretimizde bozulmalar meydana gelmiştir.

Sistem-1 içinde iki tane sık örnekleme işlemi bulundurmaktadır. Bunları adım adım yapıyoruz. Sistem-1 de dikkat etmemiz gereken x[n] işaretini ayrı ayrı sık örneklemeye sokmamak yani ilk örneklemeden çıkan işaretimizi ikinci örneklemeye sokmamız gerekiyor. Oluşturacağımız algoritma aşağıdaki sırayla yapılacaktır;
Birinci örnekleme işlemimiz;
    
    x[n] işareti L=2 ile sık örneklenmeye girer ve x_u1[n] elde edilir.
         (x_u1[n], x[n]'in örneklerinin arasına 2-1=1 örnek konmasıyla elde edilir.)
    x_u1[n], h1[n] ile lineer interpolasyon işlemine tabi tutulur ve x_i1[n] elde edilir.
   
İkinci örnekleme işlemimiz;

    x_i1[n] işareti L=3 ile sık örneklemeye girer ve x_u2[n]elde edilir.
         (x_u2[n], x_i1[n]'in örneklerinin arasına 3-1=2 örnek konmasıyla elde edilir.)
    x_u2[n], h2[n] ile lineer interpolasyon işlemine tabi tutulur ve Sistem-1 çıkışı yn_1 = y1[n] elde edilir.
    

```python
#gerekli kütüphanelerin çağrılması
from matplotlib import pyplot as plt  
import numpy as np
%matplotlib notebook
```

```python
#x[n] işaretinin tanımlanması
N=7                #x[n]'in örnek sayısı
n=np.arange(0,7)   #x[n] işaretimizin indis değerleri
xn=np.array([2,3.2,4,-5,-6.7,7.8,15]) #x[n] işaretimizin array şeklinde tanımlanması
```

     Sistem 1 içindeki 'Birinci SIK ÖRNEKLEME'nin yapılması L=2 için

```python
L1=2                       #birinci sık örnekleme kat sayımız
Nn_u1=N*L1                           #sık örnekleme yapılmış işaretimizin indis değerlerinin tanımlanması
xn_u1=np.zeros(Nn_u1)                #yeni indis değerleri sayısı kadar sıfırla dolu array tanımladık
xn_u1[np.arange(0,len(xn_u1),L1)]=xn #sıfırların üzerine L1=2 aralıklarla x[n] işaretinin değerlerini sırasıyla atadık

n_u1=np.arange(0,Nn_u1)    #örneklenmiş işaretimizin indis dizisi yazdırdık.
```

```python
#Lineer interpolasyon işleminin yapılması

hn1=np.array([0,1/2,1,1/2,0])  #L=2 için h[n] işaretini array şeklinde oluşturduk

xn_i1=np.convolve(xn_u1,hn1,'same') #Konvolüsyon işlemi ile xn_i1[n] elde edildi.
n_i1=np.arange(0,len(xn_i1))
```

    Sistem 1 içindeki 'İkinci SIK ÖRNEKLEME'nin yapılması L=3 için

```python
L2=3                    #ikinci sık örnekleme kat sayımız
Nn_u2=Nn_u1*L2                            #sık örnekleme yapılmış işaretimizin indis değerlerinin tanımlanması
xn_u2=np.zeros(Nn_u2)                     #yeni indis değerleri sayısı kadar sıfırla dolu array tanımladık
xn_u2[np.arange(0,len(xn_u2),L2)]=xn_i1   #sıfırların üzerine L2=3 aralıklarla xn_i1[n] işaretinin değerlerini sırasıyla atadık

n_u2=np.arange(0,Nn_u2)  #örneklenmiş işaretimizin indis dizisi yazdırdık.
```

    Sistem-1 çıkışının elde edilmesi

```python
#Lineer interpolasyon işleminin yapılması

hn2=np.array([0,1/3,2/3,1,2/3,1/3,0])  #L=3 için h[n] işaretini array şeklinde oluşturduk

yn_1=np.convolve(xn_u2,hn2,'same')     #Konvolüsyon işlemi ile Sistem-1 çıkışımız y1[n] elde edildi.
n_i2=np.arange(0,len(yn_1))
```
