 ## Creating a Discrete Time Signal

```python
n=np.arange (0,10) #10 noktalı olacak sekilde vektör oluşturduk
x= np.cos (n*np.pi/3+np.pi/2)
print(x)
```

## Graph plotting

```python
from matplotlib import pyplot as plt # grafik çizdirmek için pyplot komutu çağıracağız
```

```python
%matplotlib notebook  
n=np.arange(0,10)     
x=np.cos(n*np.pi/3+np.pi/2)
print(x)

plt.xlabel('n değerleri') #x-ekseni için metin yazılır 
plt.ylabel('x[n]')        #y-ekseni için metin yazılır
plt.title(r'$x[n]=\cos(\pi/3+\pi/2)$')      # bu komut ile çıktımıza başlık koyabiliriz
plt.stem(n,x,use_line_collection=True) #x değerlerinin her n değeri için aldığı değerleri çizdiririz
plt.show()
```

## Creating a Continuous Time Signal

```python
%matplotlib notebook

t=np.linspace(0,2/50,1000)     #linspaceyi genelde sürekli zamanlı sinyaller için kullanıyoruz
x=np.cos(100*np.pi*t-np.pi/5)

plt.plot(t,x)        
plt.xlabel('zaman')
plt.ylabel('x(t)')
plt.title(r'$x(t)=\cos(100\pi t-\pi/5)$')
plt.show()
```

## Plotting more than one sign in a single figure

```python
%matplotlib notebook

t=np.linspace(0,2/50,1000)

x1=np.cos(100*np.pi*t-np.pi/5)
x2=2*np.sin(200*np.pi*t)

plt.plot(t,x1,t,x2)

plt.legend(('x1(t)','x2(t)')) #grafiklere renk verir ve isimlerini kenarda yazdırır
plt.xlabel('zaman')
plt.ylabel('x(t)')

plt.show()
```

## Signal Shifting 

```python
%matplotlib notebook

n=np.array([0,1,2,3,4])
x3=np.array([1,0,2,-1,0])

plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(n,x3,use_line_collection=True) #her n değerinde x e karşılık gelen elmanı yazdırır
plt.show()
```

```python
#yukarıdaki işaretin iki birim sapa otelenmiş hali olan x[n-2] yi çizdirleim 
#bunun için iki yol vardır
```

```python
# Yol-1
%matplotlib notebook

n=np.array([0,1,2,3,4])
n2=n+2         #x[n-2] yi çizdirmek için n değerlerini 2 birim sağa telememiz gerekir

x3=np.array([1,0,2,-1,0])

plt.xlabel('n2')
plt.ylabel('x[n-2]')
plt.stem(n2,x3,use_line_collection=True)
plt.title('x[n] ifadesinin ötelenmesiyle elde edilen x[n-2]')
plt.show()
```

```python
#Yol-2
%matplotlib notebook

n=np.array([0,1,2,3,4,5])
x=np.array([1,0,2,-1,0,0])

x[n-2]=np.append(np.array([0,0]),x[:-2]) 
#append(x1,x2) fonksiyonu x1 vektörünün sonuna x2 vektörünün eklenmiş halini elde ederiz
#x[:-2] ise x vektörünün sondan iki elemanın vektörden çıkarılmış halini döndürür

fig = plt.figure()   # yeni figür elde edebilmek için 

ax1=fig.add_subplot(211)  #subplot(211) = figürü 2 satıra 1 sütuna ayırır ax1 1. bölüme çizilir.
ax1.stem(n,x,use_line_collection=True)
ax1.title.set_text('x[n]')

ax2=fig.add_subplot(212)
ax2.stem(n,x[n-2],use_line_collection=True)
ax2.title.set_text('x[n-2]')
```

```python
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib notebook
import cmath

import warnings
warnings.simplefilter("ignore", np.ComplexWarning)
```

```python
plt.figure()
r=1
theta=np.pi/4
n=np.arange(0,10)
x=r*np.exp(1j*theta*n)

plt.stem(n,x,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.title(r'$x1[n]=e^{(j\pi/4)n}$')
plt.show()
```

```python

r=1
theta=np.pi/4
n=np.arange(0,15)
x=r*np.exp(1j*theta*n)
genlik=np.abs(x)
faz=np.angle(x)

fig = plt.figure()
print('Faz değerleri:',faz)
print('')
print('Genlik değerleri:',genlik)

x1=fig.add_subplot(211)
x1.stem(n,genlik,use_line_collection=True)
x1.title.set_text('genlik')
plt.ylabel('genlik değerleri')


x2=fig.add_subplot(212)
x2.stem(n,faz,use_line_collection=True)
x2.title.set_text('faz')

plt.xlabel('zaman')
plt.ylabel('faz değerleri')

plt.show()
```

```python
plt.figure()
theta=1j*np.pi/4

n=np.arange(-5,5,1)


x = np.zeros(len(n))    # sinyali depolamak için bellekte n (vektörü) uzunluğunda bir yer ayır

for k in range(len(n)): # her k değeri için x[k]değeri aynı çıkar çünkü işaretimiz n değerine bağlı değildir.
    x[k]=np.exp(theta)
    
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.title(r'$x2[n]=e^{j\pi/4}$' )
plt.stem(n,x,use_line_collection=True)
plt.show()
```

```python
plt.figure()
theta=np.pi/4
n=np.arange(-5,5,1)

x = np.zeros(len(n)) # sinyali depolamak için bellekte n (vektörü) uzunluğunda bir yer ayır

for k in range(len(n)): # her n değeri için x[n]değeri aynı çıkar çünkü iişaretimiz n değerine bağlı değil
    x[k]=np.exp(theta)
    
plt.xlabel('n')
plt.ylabel('x3[n]')
plt.title(r'$x3[n]=e^{\pi/4}$')
plt.stem(n,x,use_line_collection=True)
plt.show()

```
